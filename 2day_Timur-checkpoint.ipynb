### Визуализируем данные на Python



```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Загрузим датасет Ирис (подробнее о датасете https://ru.wikipedia.org/wiki/%D0%98%D1%80%D0%B8%D1%81%D1%8B_%D0%A4%D0%B8%D1%88%D0%B5%D1%80%D0%B0)
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris.head()
```

### 1. Постройте точечный график
Создайте точечный график (scatterplot), где по оси X - sepal_length, а по Y - sepal_width <br>
Добавьте название графика и название осей

```python
x=iris['sepal_length']
y=iris['sepal_width']
plt.scatter(x,y)
plt.gca().set(xlabel='sepal_length', ylabel='sepal_width')
plt.show
```

### 2. Постройте BarPlot
Создайте BarPlot, который покажет частоты по всем видам Ириса

```python
iris_new=iris.copy()
#pd.get_dummies(iris_new['species'])
iris_new=pd.concat([iris_new, pd.get_dummies(iris_new['species'])], axis=1)
setosa=iris_new['setosa'].sum()
versicolor=iris_new['versicolor'].sum()
virginica=iris_new['virginica'].sum()
y=[setosa,versicolor,virginica]
x=['setosa','versicolor','virginica']
fig, ax = plt.subplots()
ax.bar(x, y)
plt.show()
```

### 3. Постройте PieChart (круговая диаграмма)
Создайте PieChart для видов Ириса

```python
fig, ax = plt.subplots()
ax.pie(y,labels=x)
plt.show()
```
